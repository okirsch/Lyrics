### Texte

Ich suche die Liedertexte die möglichst von einem Songwriter geschrieben wurden. Danach trainiere ich die Modelle, die den Autor nach dem Liedertext erkennen können.

```python
page = requests.get('https://www.lyrics.com/lyric/29598234/Everlast/Weakness')
ever = page.text
everlast_soup = BeautifulSoup(ever, 'html.parser')
everlast = everlast_soup.find('pre', attrs = {'id' : 'lyric-body-text'}).text.replace("\'", " ").replace("\n", " ")
everlast
```

### Funktion

```python
"""Funktion, die mir einen klaren Text abruft"""
def gib_text(link):
    page = requests.get(link)
    soup = BeautifulSoup(page.text, 'html.parser')
    text = soup.find('pre', attrs = {'id' : 'lyric-body-text'}).text.replace("\'", " ").replace("\n", " ")
    return text
```

```python
everlast1 = gib_text('https://www.lyrics.com/lyric/29598235/Everlast/Stone+in+My+Hand')
everlast1
```

```python
everlast2 = gib_text('https://www.lyrics.com/lyric/29598232/Everlast/Stay')
#everlast2
```

```python
everlast3 = gib_text('https://www.lyrics.com/lyric/14925415/Everlast/Let+It+Go')
#everlast3
```

```python
"""LP- Laura Pergolizzi schreibt oft ihre Texte zusammen mit anderen Autoren, ich nehme möglichst die gleichen"""

lp = gib_text('https://www.lyrics.com/lyric/30783186/LP/Forever+for+Now')# Laura Pergolizzi
#lp
```

```python
lp1 = gib_text('https://www.lyrics.com/lyric/33551266/LP/Lost+on+You')#Laura Pergolizzi, Nathaniel Campany, Michael Gonzales
#lp1
```

```python
lp2 = gib_text('https://www.lyrics.com/lyric/33972293/LP/When+We%27re+High')#Laura Pergolizzi, Nathaniel Campany, Mike del Rio
#lp2
```

```python
lp3 = gib_text('https://www.lyrics.com/lyric/35768667/LP/Girls+Go+Wild')# Laura Pergolizzi, Michael Gonzalez, Nathaniel Campany
#lp3
```

```python
lp4 = gib_text('https://www.lyrics.com/lyric/33925842/LP/Other+People') #  Laura Pergolizzi, Nathaniel Campany, Benjamin Joseph Romans
lp4
```

### Build a Lyrics Corpus

```python
X_train = [everlast, everlast1, everlast2, everlast3, lp, lp1, lp2, lp3]
X_test = [lp4] 
y_train = ['Everlast', 'Everlast', 'Everlast', 'Everlast', 'LP', 'LP', 'LP', 'LP']
y_test = ['LP']
```

### Vectorizing
#### Stop_words



```python
from sklearn.feature_extraction.text import CountVectorizer# konventiere Text in eine Matrix
```

```python
cv = CountVectorizer(stop_words='english') 
cv
```

```python
cv.fit(X_train)                    
out_train = cv.transform(X_train)

out_test = cv.transform(X_test)
```

```python
out_train.todense()
out_test.todense()
out_test.shape
```

#### Sparse matrix

```python
df = pd.DataFrame(out_train.todense(), columns = cv.get_feature_names())
df
```

```python
df_test = pd.DataFrame(out_test.todense(), columns = cv.get_feature_names())
df_test
```

## Tf-Idf (Normalization)

```python
from sklearn.feature_extraction.text import TfidfTransformer #Ziel - die Auswirkungen von Token zu verringern, die in einem bestimmten Korpus sehr häufig auftreten 
```

```python
tf = TfidfTransformer()
transformed_train = tf.fit_transform(out_train)

tdf = pd.DataFrame(transformed_train.todense(), columns = cv.get_feature_names())
tdf
```

```python
transformed_test = tf.fit_transform(out_test)
tdf_test = pd.DataFrame(transformed_test.todense(), columns = cv.get_feature_names())
tdf_test
```

### Naive Bayes Classifier

```python
from sklearn.naive_bayes import MultinomialNB
X_tr = tdf.values
X_te = tdf_test.values
```

```python
m = MultinomialNB(alpha = 0.01)
m.fit(X_tr,y_train)
m.score(X_tr,y_train)
```

```python
m.predict_proba(X_tr).round(5)
```

```python
y_pred = m.predict(X_tr)
y_pred_test = m.predict(X_te) #als Test ist Text von LP


m.predict_proba(X_te).round(3)
print(y_pred_test)
```

```python
print(y_pred_test)
print('Text definiert als Text von ' + str(y_pred_test))
```
